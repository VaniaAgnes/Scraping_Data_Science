```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd

# Downloading Rotten Tomatoes data
headers = {'Accept-Language': 'en-US,en;q=0.8'}
url = 'https://editorial.rottentomatoes.com/guide/popular-tv-shows/'
response = requests.get(url,headers=headers)
soup = BeautifulSoup(response.text, "html.parser")

#Scrape movie title
# movietitle = soup.select('h2 a')

movietitle=[]
links=[]
year=[]
score=[]
actors=[]

for t, b, c, a in zip(soup.select('h2 a'), soup.select('span.subtle'), soup.select('span.tMeterScore'), soup.select('div.info a')):
  movietitle.append(t.get_text())
  links.append(t.attrs.get('href'))
  year.append(b.get_text())
  score.append(c.get_text())
  actors.append(a.get_text())


# #Scrap Movie Year
# for b in soup.select('span.subtle'):
#   year.append(t.get_text())

print(movietitle)
print(links)
print(year)
print(score)
print(actors)

#Scrape movie years
year = soup.select('span.subtle')

# Ensure all lists have the same length by padding with empty strings
max_length = max(len(movietitle), len(links), len(year), len(score), len(actors))
movietitle += [''] * (max_length - len(movietitle))
links += [''] * (max_length - len(links))
year += [''] * (max_length - len(year))
score += [''] * (max_length -len(score))
actors += [''] * (max_length -len(actors))


df = pd.DataFrame({
    'Movie Title': movietitle,
    'Link': links,
    'Year': year, 'Score': score, 'Actors': actors

})

# Print the first few rows of the DataFrame
print(df.head())

df.to_csv('rottentomatoes.csv', index=False)

```

```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd

# Downloading MyDramaList data
headers = {'Accept-Language': 'en-US,en;q=0.8'}
url = 'https://mydramalist.com/shows/top_korean_dramas'
response = requests.get(url,headers=headers)
soup = BeautifulSoup(response.text, "html.parser")

movietitles=[]
links=[]
ranking=[]
info=[]
score=[]

#Data Looping
for a in soup.select('h6.title a'):
    movietitle = a.get_text() if a else ''
    link = a.attrs.get('href') if a else ''

    if movietitle.strip() != '' and link.strip() != '':
        movietitles.append(movietitle)
        links.append(link)

for b in soup.select('div.ranking'):
  ranking.append(b.get_text())

for c in soup.select('span.text-muted'):
  info.append(c.get_text())

for d in soup.select('span.score'):
  score.append(d.get_text())



#Printing the Data
print(movietitles)
print(links)
print(ranking)
print(info)
print(score)


#Ensuring Array Length is the Same
max_length = max(len(movietitles), len(links), len(ranking), len(info), len(score))
movietitles += [''] * (max_length - len(movietitles))
links += [''] * (max_length - len(links))
ranking += [''] * (max_length - len(ranking))
info += [''] * (max_length -len(info))
score += [''] *  (max_length - len(score))



df = pd.DataFrame({
    'Movie Title': movietitles,
    'Ranking': ranking,
    'Links': links, "Information": info, 'Score': score
})

# Print the first few rows of the DataFrame
print(df.head())

df.to_csv('mydramalist.csv', index=False)







```
